```python
""" 패키지 불러오기 """
import pandas as pd
```

```python
""" 데이터 불러오기 """
krx_stock_info = pd.read_excel('krx_stock_info.xlsx', engine="openpyxl")
stock_n_candidate_name = pd.read_csv('stock_n_candidate_name.csv')
```

```python
""" 테이블 병합 """

# 필요한 종목이 있는 테이블에 LEFT JOIN
merged_df = stock_n_candidate_name.merge(
    krx_stock_info,
    left_on="stock_name",
    right_on="한글 종목약명",
    how="left"  # 일치하는 데이터만 가져옴
)
```
